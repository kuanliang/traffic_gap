```python
import pandas as pd
import numpy as np
```

```python
ewma = pd.stats.moments.ewma
```

```python
x = np.linspace(0, 1, 100)
x = np.hstack((x,x[::-1]))
```

```python
x += np.random.normal(loc=0, scale=0.1, size=200)
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
plt.plot(x, alpha=0.4, label='Raw')
```

```python
fwd = ewma(x, span=15)
```

```python
bwd = ewma(x[::-1], span=15)
```

```python
c = np.vstack((fwd, bwd[::-1]))
```

```python
c = np.mean(c, axis=0)
```

```python
c
```

```python
len(c)
```

```python
plt.plot(x, alpha=0.4, label='Raw')
plt.plot(ewma(x, span=20), 'b', label='EWMA, space=20')
plt.plot(c, 'r', label='Reversed-Recombined')
#legend(loc=8)
```
